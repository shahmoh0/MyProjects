```python
import pandas as pd
import time
import copy
```

```python
df = pd.read_csv("clean_passenger_id.csv")

data = df["Passenger ID"].tolist()

print(data[:10])
```

```python

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key

    return arr
```

```python
print(insertion_sort([5, 3, 1, 4]))
```

```python
def average_time(arr, repetitions=10):
    total_time = 0

    for _ in range(repetitions):
        temp = arr.copy()

        start = time.perf_counter()
        insertion_sort(temp)
        end = time.perf_counter()

        total_time += (end - start)

    average = total_time / repetitions
    return average * 1000
```
